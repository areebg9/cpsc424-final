```python
import numpy as np
import matplotlib.pyplot as plt
# One-shot results (redo with averages later)
# X-axis: number of bodies
x = [1000, 10000, 100000, 200000, 500000, 1000000]
# Y-axis: execution time in seconds
y_brute_force_seq = [0.001761, 0.176571, 17.5902, 70.3169]
y_brute_force_omp = [0.020762, 0.047675, 2.47074, 9.49367, 64.8364, 301.41]
y_brute_force_parlay = [0.001227, 0.024268, 2.07556, 8.43054, 58.5038, 258.817]

plt.figure(figsize=(10, 6))
plt.plot(x[:4], y_brute_force_seq, "o-", label="Brute Force Sequential")
plt.plot(x, y_brute_force_omp, "s-", label="Brute Force OpenMP")
plt.plot(x, y_brute_force_parlay, "^-", label="Brute Force Parlay")

plt.xscale("log")
plt.yscale("log")
plt.xlabel("Number of Bodies (log scale)")
plt.ylabel("Execution Time (s, log scale)")
plt.title("N-Body Simulation Runtime Comparison")
plt.legend()
plt.grid(True, which="both", linestyle="--", linewidth=0.5)
plt.tight_layout()
plt.show()

speedup = np.array(y_brute_force_omp[1:]) / np.array(y_brute_force_parlay[1:])
plt.figure(figsize=(8, 5))
plt.plot(x[1:], speedup, "d-", color="purple")
plt.xscale("log")
plt.xlabel("Number of Bodies (log scale)")
plt.ylabel("OpenMP / Parlay Speedup Ratio")
plt.title("Parlay Speedup over OpenMP")
plt.grid(True)
plt.axhline(1, color="gray", linestyle="--")
plt.tight_layout()
plt.show()
```
